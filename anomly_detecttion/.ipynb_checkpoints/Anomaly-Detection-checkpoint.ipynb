```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,8)
```

```python
from sklearn.neighbors import LocalOutlierFactor
from sklearn.ensemble import IsolationForest
from sklearn.covariance import EllipticEnvelope
```

```python
df = pd.read_csv('AnomalyDetection.csv')
df.shape
```

```python
df.head()
```

```python
df['Mileage'] += 10
```

```python
plt.scatter(df['Mileage'],df['Price'], s=5)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

# Elliptic Envelop

```python
model = EllipticEnvelope(contamination=0.01)
ee = model.fit_predict(df.iloc[:, :2])
df['is_outlier_ee'] = ee
```

```python
df['is_outlier_ee'].value_counts()
```

```python
plt.scatter(df['Mileage'],df['Price'], s=10, c=df['is_outlier_ee'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
less_dense = df.loc[df['Mileage'] <= 80]
```

```python
label = EllipticEnvelope(random_state=0, contamination=0.04).fit_predict(less_dense.iloc[:,:2])
less_dense['is_outlier_ee'] = label

# Plotting
plt.scatter(less_dense['Mileage'], less_dense['Price'], c=less_dense['is_outlier_ee'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
less_dense['Mileage'].mean(), less_dense['Price'].mean()
```

```python
less_dense['Mileage'].std(), less_dense['Price'].std()
```

```python
np.cov(less_dense['Mileage'], less_dense['Price'])
```

# Isolation Forest

```python
# clf = IsolationForest(contamination = 0.05, random_state=0, n_estimators=1000)
```

```python
label = IsolationForest(contamination = 0.04, random_state=0, n_estimators=10).fit_predict(df.iloc[:, :2])


df['is_outlier_if'] = label
plt.scatter(df['Mileage'], df['Price'], s=10, c=df['is_outlier_if'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

# LOF

```python
model = LocalOutlierFactor(n_neighbors=25, contamination=0.04)
label = model.fit_predict(df.iloc[:, :2])

df['is_outlier_lof'] = label

plt.scatter(df['Mileage'], df['Price'], s=10, c=df['is_outlier_lof'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.show()
```

```python
df['is_outlier_lof'].value_counts()
```
